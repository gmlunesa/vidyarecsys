```python
# Import necessary libraries
import pandas
import pylab as pl

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Read the data
df = pandas.read_csv('matrixdata.csv',
                    index_col = 0)

Y = df[['action']]
Z = df[['strategy']]
A = df[['rpg']]
B = df[['indie']]
C = df[['adventure']]
D = df[['sports']]
E = df[['simulation']]
F = df[['mmo']]
G = df[['free']]
H = df[['casual']]

df.index.names = ['id']
df.columns.names = ['genre']
```

```python
# Dimensionality reduction
pca = PCA(copy=True, n_components=3, whiten=False)
pca.fit(df)

df_2d = pca.transform(df)
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(Y).score(Y) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for Action Items')
pl.show()

score = [kmeans[i].fit(Z).score(Z) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for Strategy Items')
pl.show()

score = [kmeans[i].fit(A).score(A) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for RPG Items')
pl.show()


score = [kmeans[i].fit(B).score(B) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for Indie Items')
pl.show()

score = [kmeans[i].fit(C).score(C) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for Adventure Items')
pl.show()

score = [kmeans[i].fit(D).score(D) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for Sports Items')
pl.show()

score = [kmeans[i].fit(E).score(E) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for Simulation Items')
pl.show()


score = [kmeans[i].fit(F).score(F) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for MMO Items')
pl.show()

score = [kmeans[i].fit(G).score(G) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for Free Items')
pl.show()

score = [kmeans[i].fit(H).score(H) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for Casual Items')
pl.show()
```
