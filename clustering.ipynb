```python
import pandas
import pylab as pl

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df = pandas.read_csv('matrixdata.csv',
                    index_col = 0)

Y = df[['action']]
Z = df[['strategy']]
A = df[['rpg']]
B = df[['indie']]
C = df[['adventure']]
D = df[['sports']]
E = df[['simulation']]
F = df[['mmo']]
G = df[['free']]
H = df[['casual']]

df.index.names = ['id']
df.columns.names = ['genre']

print(df)
```

```python
# Dimensionality reduction
pca = PCA(copy=True, n_components=3, whiten=False)
pca.fit(df)

df_2d = pca.transform(df)
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(Y).score(Y) for i in range(len(kmeans))]
score
pl.plot(Nc,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve for Action Items')
pl.show()
```

```python
import pandas as pd
from scipy import stats
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

clmns = ['action', 'strategy','rpg', 'indie', 'adventure', 'sports', 'simulation', 'mmo', 'free', 'casual']
df_tr_std = stats.zscore(df[clmns])
```

```python
kmeans = KMeans(n_clusters=6, random_state=0).fit(df_tr_std)
labels = kmeans.labels_

df['clusters'] = labels

clmns.extend(['clusters'])

sns.lmplot('action', 'strategy', 
           data=df, 
           fit_reg=False, 
           hue="clusters",  
           scatter_kws={"marker": "D", 
                        "s": 100})
plt.title('Clusters Action vs Strategy')
plt.xlabel('action')
plt.ylabel('strategy')
```
